- We want to use a house's features (e.g. square footage, bathrooms, etc.)
    - We want to use these features to predict the sales price of the house

**What is the goal, and how might you naively address it?**

- I want to know the value of my house

- I could look at recent sales in my neighborhood to estimate what I think mine is worth
    - Let's say I record 1) the sales price and 2) the size of the house
    
![](images/1.PNG)

**Terminology**

- The x-variable is called the **feature** or **predictor**
- The y-variable is called the **observation** or **response variable**

**Linear Regression**

- We fit a line to the data

![](images/2.PNG)

- $w_{0}$ and $w_{1}$ are called the **parameters** of our model
    - That's why we'll call the model $f_{w}(x)$, where $w=(w_{0},w_{1})$

**How did we fit the line?**

- We fit it using the **residual sum of squares (RSS)**
    - We loop through the different possible values of $w_{0}$ and $w_{1}$ and minimized the squared error

**How do we use the line to predict the value of the house?**

- We can just look for the value of the line at our square footage

![](images/3.PNG)

**Is our fitted line sufficient?**

- Only if it's a linear relationship

- But what if it's a quadratic relationship?

### Note: even though this is a quadratic function, it is still called a linear regression

**Why?**

- Because we have **one** independent variable and **one** dependent variable

**But we can keep making our models more and more complicated to minimize the RSS (residuals)**

- How do we know when to stop? And how can we test which model is better?

### Overfitting

- When a model fits the data too closely
    - The model fits the idiosyncracies of the data so well that it won't work well at making predictions
    
- **How do we test for overfitting?**
    - We split the historical data into 1) a **training set** and 2) a **testing set**
        - We fit the model to the training set and use it to make predictions
            - We then compare our predictions to the values in the testing set to see how accurate it is
            
- When performing the **training testing split**, we fit the model to the the training data by **minimizing the training error**

- When we're comparing our predicted values to the testing data points, we calculate the **testing error**

### Training/Test Curves

- For each model we generate from the testing data, we calculate the training error and testing error generated by the model
    - We can then plot the relationship between **model complexity** and the **training error**, and/or between the **model complexity** and the **testing error**
